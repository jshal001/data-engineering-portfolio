```python
schema_name = "spotify_pipeline"
volume_name = "spotify_tracks"

spark.sql(f"CREATE SCHEMA IF NOT EXISTS workspace.{schema_name}")
spark.sql(f"CREATE VOLUME IF NOT EXISTS workspace.{schema_name}.{volume_name}")
```

```python
spark.sql(f"DROP TABLE IF EXISTS workspace.{schema_name}.spotifytracks_bronze")

spark.sql(f"""
    CREATE TABLE IF NOT EXISTS workspace.{schema_name}.spotifytracks_bronze
    (
        _c0 STRING,
        track_id STRING,
        artists STRING,
        album_name STRING,
        track_name STRING,
        popularity STRING,
        duration_ms STRING,
        explicit STRING,
        danceability STRING,
        energy STRING,
        key STRING,
        loudness STRING,
        mode STRING,
        speechiness STRING,
        acousticness STRING,
        instrumentalness STRING,
        liveness STRING,
        valence STRING,
        tempo STRING,
        time_signature STRING,
        track_genre STRING
    )
""").display()
```

```python
spark.sql(f'''
          COPY INTO workspace.{schema_name}.spotifytracks_bronze FROM '/Volumes/workspace/{schema_name}/{volume_name}'
          FILEFORMAT = CSV
          FORMAT_OPTIONS ('header' = 'true', 'quote' = '"', 'escape' = '"')
          ''').display()

```
